```python
dev = 0.008
bdict = {}
tdict = {}
for ti,ts,tdc,td,tw in dfc2_t.itertuples():
    x1 = tw-dev*tw
    x2 = tw+dev*tw
    for bi,bs,bdc,bd,bw in dfc2_b.itertuples():
        y1 = bw-dev*bw
        y2 = bw+dev*bw
        if ((x1<=bw) and (bw<=x2))or((y1<=tw) and (tw<=y2)):
            print td + " are similar " + bd
            print "%d worker %d"%(tw,bw)
            bdict[bd] = bw
            tdict[td] = tw
```

```python
for k,v in bdict.iteritems():
    print k 
    print v
```

```python
n_groups = 9 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
rects1 = plt.bar(index, tdict, bar_width,
                alpha=opacity,
                color='b')

rects2 = plt.bar(index + bar_width, bdict, bar_width,
                alpha=opacity,
                color='g')
plt.show()
```
