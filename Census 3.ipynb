```python
import numpy as np                # to load numpy library
import pandas as pd               # to load panda library
import matplotlib.pyplot as plt   # to load the matplotlib library only pyplot to ploy the graphs
%matplotlib inline                
%pylab inline
```

```python
Location = r'data/india-districts-census-2011.csv'  # location where the data is stored
df = pd.read_csv(Location)                          # reading the file from the location and storing it in the dataframes
```

```python
#list(df)
```

```python
dfc3 = df[["State name","District name","Workers","Households","Agricultural_Workers","Households_with_Telephone_Mobile_Phone"]]
```

```python
dfc3b = dfc3[(dfc3["Agricultural_Workers"]/dfc3["Workers"]*100 > 50)]
dfc3b
```

```python
dfc3e = dfc3b.groupby("State name").sum()
dfc3e
```

```python
dfc3f = dfc3e["Households_with_Telephone_Mobile_Phone"]/dfc3e["Households"]*100
```

```python
dfc3f.head(dfc3f.count()).plot(kind='bar')
plt.show()
```

```python
dfc3g = dfc3[(50 > dfc3["Agricultural_Workers"]/dfc3["Workers"]*100)].groupby("State name").sum()
dfc3g
dfc3h = dfc3g["Households_with_Telephone_Mobile_Phone"]/dfc3g["Households"]*100
dfc3h
dfc3h.plot(kind='bar')
plt.show()
```

```python
dfc3e = dfc3b[(dfc3b['State name'] == 'BIHAR')]
dfc3e = dfc3e.set_index("District name")
dfc3e
dfc3f = dfc3e["Households_with_Telephone_Mobile_Phone"]/dfc3e["Households"]*100
dfc3f.plot(kind='bar')
plt.show()
```
